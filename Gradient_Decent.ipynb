# Trigonometric Ratio Table
Tangent 90 and 270 degrees are underfined



<img src="https://physicscatalyst.com/article/wp-content/uploads/2019/01/trigonometric-ratios-tables-1.png">

```python
import matplotlib.pyplot as plt
import numpy as np
import math
plt.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(figsize=(20, 8))
tan_all = []
angles = []
# for angle in range (0, 361, 10):
for angle in range (-360, 361, 20):
    if angle == 90 or angle == 270 or angle == -90 or angle == -270:
        continue
    slope = math.tan(angle * math.pi / 180)
    tan_all.append(slope)
    angles.append(angle)
ax.set_ylabel('Slope Value', fontsize=12)
ax.set_xlabel('Slope Angle', fontsize=12)
ax.plot( tan_all) 
ax.set_xticks(np.arange(len(angles)))
ax.xaxis.grid(True)
ax.set_xticklabels(angles, fontsize=10);
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
# plt.rcParams.update({'font.size': 6})
fig, ax = plt.subplots(figsize=(20, 8))
angle_all = []
val_all = []
for idx, val in enumerate(np.arange(-10, 10, 0.5)):
    angle = math.degrees(math.atan(val))
    angle_all.append(angle)
    val_all.append(val)

ax.set_ylabel('Slope Angle', fontsize=12)
ax.set_xlabel('Slope Value', fontsize=12)
ax.plot(angle_all) 
ax.set_xticks(np.arange(len(val_all)))
ax.xaxis.grid(True)
ax.set_xticklabels(val_all, fontsize=9);
```


# Gradient Descent Example
src: https://towardsdatascience.com/gradient-descent-unraveled-3274c895d12d

<img src="https://miro.medium.com/max/640/0*bDbmkA9XIiu28SNo">

```python
import numpy as np
# Define the equation function
def loss_surface(x):
    return np.power(x,4)+ np.power(x,3) - 6*np.power(x,2)
#Define the gradient function
def parabola_grad(x):
    return 4*np.power(x,3)+3*np.power(x,2)-12*x
# Select a random start point in the axis range
x0 = np.random.uniform(-3,3)
# Specify the learning rate
lr=0.001
 
# Initialize a list to keep track of the gradient descent points
x_l = [x0]
#Perform gradient descent
for i in range(10000):
    # Update the x0 value 
    x0=x0-parabola_grad(x0)*lr
    # Add points to the list
    x_l.append(x0)

 
x_l = np.array(x_l)    
y_l = loss_surface(x_l)
# Plot the curve and points

x = np.arange(-3, 3, 0.01)
y = loss_surface(x)  
plt.plot(x,y, label='convex equation');
plt.plot(x_l[0],y_l[0],'rx', label='random start point');
plt.plot(x_l[-1],y_l[-1],'kx', label='minima point');
plt.plot(x_l,y_l,'r-', label='gradient descent');
plt.legend()
```
